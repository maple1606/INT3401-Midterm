```python
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()

DATAPATH = os.getenv('DATAPATH')


data = pd.read_csv(DATAPATH + "weather_data_4.csv")
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d %H:%M:%S')
```

```python
train_df = data[(data['date'] >= f'2019-{4}-05 00:00:00') & (data['date'] < f'2020-{4}-28')].reset_index(drop=True)
train_df.head()
```

```python
test_df = data[(data['date'] < f'2019-{4}-3') & (data['date'] >= f'2020-{4}-28 00:00:00')].reset_index(drop=True)
test_df.shape
```
